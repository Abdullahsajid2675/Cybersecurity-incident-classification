```python
numeric_cols = [
    'data_compromised_GB',
    'attack_duration_min',
    'attack_severity',
    'response_time_min'
]
iqr_summary = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()

    iqr_summary[col] = outliers

print(iqr_summary)
```

Feature Engineering on the Timestamp Feature

```python
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")

df = df.dropna(subset=["timestamp"])

df = df.sort_values("timestamp").reset_index(drop=True)

df["year"]        = df["timestamp"].dt.year.astype("int16")
df["month"]       = df["timestamp"].dt.month.astype("int8")
df["day"]         = df["timestamp"].dt.day.astype("int8")
df["day_of_week"] = df["timestamp"].dt.weekday.astype("int8")  # 0=Mon, 6=Sun
df["hour"]        = df["timestamp"].dt.hour.astype("int8")
df["minute"]      = df["timestamp"].dt.minute.astype("int8")

#binary behavioral flags
df["is_weekend"] = (df["day_of_week"] >= 5).astype("int8")
df["is_night"]   = ((df["hour"] >= 22) | (df["hour"] <= 5)).astype("int8")
df["is_business_hours"] = ((df["hour"] >= 9) & (df["hour"] <= 17)).astype("int8")


#Cyclical Encoding (Neural Net Critical)
import numpy as np
df["hour_sin"]  = np.sin(2 * np.pi * df["hour"] / 24)
df["hour_cos"]  = np.cos(2 * np.pi * df["hour"] / 24)

df["dow_sin"]   = np.sin(2 * np.pi * df["day_of_week"] / 7)
df["dow_cos"]   = np.cos(2 * np.pi * df["day_of_week"] / 7)

df["month_sin"] = np.sin(2 * np.pi * df["month"] / 12)
df["month_cos"] = np.cos(2 * np.pi * df["month"] / 12)
```

```python
# dropping redundant rows/columns
df.drop(columns=[
    "timestamp",
    "year", "month", "day",
    "hour", "minute", "day_of_week"
], inplace=True)

df.info()
```

## Manual Train-Test Split

### Subtask:
Implement a manual train-test split for `X` and `y`, dividing the data into 80% training and 20% testing sets using `numpy` for shuffling indices to ensure reproducibility.


**Reasoning**:
To implement a manual train-test split, I need to import numpy, generate and shuffle indices, then use these indices to split the X and y data into training and testing sets, and finally verify their shapes.



```python
import numpy as np

total_samples = len(X)
indices = np.arange(total_samples)
np.random.seed(42)
np.random.shuffle(indices)

split_point = int(total_samples * 0.8)

train_indices = indices[:split_point]
test_indices = indices[split_point:]

X_train, X_test = X.iloc[train_indices], X.iloc[test_indices]
y_train, y_test = y.iloc[train_indices], y.iloc[test_indices]

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

## Manual Feature Scaling (Standardization)

### Subtask:
Apply manual standardization using `numpy` to all continuous numerical features in both the training and testing sets, using statistics derived from `X_train` to prevent data leakage.


**Reasoning**:
To apply manual standardization to the specified numerical features, I will first identify these columns, then calculate the mean and standard deviation for each from `X_train`, and finally apply these statistics to scale both `X_train` and `X_test` to prevent data leakage.



```python
continuous_numerical_cols = [
    'data_compromised_GB', 'attack_severity', 'attack_duration_min', 'response_time_min',
    'attacker_ip_oct_1', 'attacker_ip_oct_2', 'attacker_ip_oct_3', 'attacker_ip_oct_4',
    'target_ip_oct_1', 'target_ip_oct_2', 'target_ip_oct_3', 'target_ip_oct_4',
    'hour_sin', 'hour_cos', 'dow_sin', 'dow_cos', 'month_sin', 'month_cos',
    'response_efficiency', 'attack_intensity'
]

# Store means and stds calculated from X_train
scaling_params = {}

for col in continuous_numerical_cols:
    mean_val = X_train[col].mean()
    std_val = X_train[col].std()
    scaling_params[col] = {'mean': mean_val, 'std': std_val}

    # Apply scaling to X_train
    X_train[col] = (X_train[col] - mean_val) / std_val

    # Apply scaling to X_test using X_train's stats
    X_test[col] = (X_test[col] - mean_val) / std_val

print("Shapes after scaling:")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")

print("\nVerification of scaling (mean and std for a few columns in X_train and X_test):")
for col in ['data_compromised_GB', 'attack_duration_min', 'response_efficiency']:
    print(f"\nColumn: {col}")
    print(f"X_train - Mean: {X_train[col].mean():.4f}, Std: {X_train[col].std():.4f}")
    print(f"X_test - Mean: {X_test[col].mean():.4f}, Std: {X_test[col].std():.4f}")
```

**Reasoning**:
The previous code produced a `SettingWithCopyWarning` because it was attempting to modify a view or copy of a DataFrame slice, which can lead to unexpected behavior. To fix this, I will explicitly use `.loc` for assignments to ensure modifications are made directly to the DataFrame and suppress the warning, while keeping the logic the same.



```python
continuous_numerical_cols = [
    'data_compromised_GB', 'attack_severity', 'attack_duration_min', 'response_time_min',
    'attacker_ip_oct_1', 'attacker_ip_oct_2', 'attacker_ip_oct_3', 'attacker_ip_oct_4',
    'target_ip_oct_1', 'target_ip_oct_2', 'target_ip_oct_3', 'target_ip_oct_4',
    'hour_sin', 'hour_cos', 'dow_sin', 'dow_cos', 'month_sin', 'month_cos',
    'response_efficiency', 'attack_intensity'
]

# Store means and stds calculated from X_train
scaling_params = {}

for col in continuous_numerical_cols:
    mean_val = X_train[col].mean()
    std_val = X_train[col].std()
    scaling_params[col] = {'mean': mean_val, 'std': std_val}

    # Apply scaling to X_train using .loc to avoid SettingWithCopyWarning
    X_train.loc[:, col] = (X_train[col] - mean_val) / std_val

    # Apply scaling to X_test using X_train's stats and .loc
    X_test.loc[:, col] = (X_test[col] - mean_val) / std_val

print("Shapes after scaling:")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")

print("\nVerification of scaling (mean and std for a few columns in X_train and X_test):")
for col in ['data_compromised_GB', 'attack_duration_min', 'response_efficiency']:
    print(f"\nColumn: {col}")
    print(f"X_train - Mean: {X_train[col].mean():.4f}, Std: {X_train[col].std():.4f}")
    print(f"X_test - Mean: {X_test[col].mean():.4f}, Std: {X_test[col].std():.4f}")
```

```python
import numpy as np

class LogisticRegressionScratch:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None

    def _sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def _cost_function(self, y_true, y_pred):
        # Binary Cross-Entropy Loss
        # Avoid log(0) by clipping probabilities
        epsilon = 1e-10
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize weights and bias
        self.weights = np.zeros((n_features, 1))
        self.bias = 0.0

        # Reshape y to be a column vector for consistent operations
        y = y.reshape(-1, 1)

        # Gradient Descent
        for _ in range(self.n_iterations):
            # Linear combination of inputs
            z = np.dot(X, self.weights) + self.bias

            # Predicted probabilities (sigmoid activation)
            y_pred = self._sigmoid(z)

            # Calculate gradients
            dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / n_samples) * np.sum(y_pred - y)

            # Update weights and bias
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict_proba(self, X):
        if self.weights is None or self.bias is None:
            raise RuntimeError("Model has not been trained yet. Call fit() first.")
        z = np.dot(X, self.weights) + self.bias
        return self._sigmoid(z)

    def predict(self, X):
        probabilities = self.predict_proba(X)
        # Convert probabilities to binary class labels (0 or 1)
        return (probabilities >= 0.5).astype(int)

print("LogisticRegressionScratch class defined successfully.")
```

## Calculate Precision, Recall, F1-Score Manually

### Subtask:
Manually calculate precision, recall, and F1-score for each class, and then compute the macro average. This involves defining helper functions for true positives, false positives, false negatives, and deriving the metrics.

#### Instructions:
1.  Define a function `calculate_binary_metrics(y_true_binary, y_pred_binary)` that takes binary true and predicted labels and returns TP, FP, FN.
2.  Iterate through each unique class in `y_test`.
3.  For each class, convert `y_test` and `y_pred_ovr` into binary arrays where the current class is 1 and all others are 0.
4.  Use `calculate_binary_metrics` to get TP, FP, FN for the current class.
5.  Calculate precision, recall, and F1-score for the current class. Handle division by zero for precision and recall.
6.  Store these metrics for each class.
7.  Compute the macro average for precision, recall, and F1-score by averaging the scores across all classes.
8.  Print the macro averaged precision, recall, and F1-score, formatted to four decimal places.

**Reasoning**:
To manually calculate precision, recall, and F1-score for each class and their macro averages, I will define helper functions to compute true positives, false positives, and false negatives, then iterate through each unique class to derive these metrics and average them.



```python
def calculate_binary_metrics(y_true_binary, y_pred_binary):
    TP = np.sum((y_true_binary == 1) & (y_pred_binary == 1))
    FP = np.sum((y_true_binary == 0) & (y_pred_binary == 1))
    FN = np.sum((y_true_binary == 1) & (y_pred_binary == 0))
    return TP, FP, FN

all_precisions = []
all_recalled_values = []
all_f1_scores = []

for class_label in unique_classes:
    # Convert y_test and y_pred_ovr into binary arrays for the current class
    y_true_binary = (y_test_np == class_label).astype(int)
    y_pred_binary = (y_pred_ovr_np == class_label).astype(int)

    TP, FP, FN = calculate_binary_metrics(y_true_binary, y_pred_binary)

    # Calculate Precision
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    all_precisions.append(precision)

    # Calculate Recall
    recalled_value = TP / (TP + FN) if (TP + FN) > 0 else 0
    all_recalled_values.append(recalled_value)

    # Calculate F1-Score
    f1_score = 2 * (precision * recalled_value) / (precision + recalled_value) if (precision + recalled_value) > 0 else 0
    all_f1_scores.append(f1_score)

# Compute macro averages
macro_precision = np.mean(all_precisions)
macro_recalled_value = np.mean(all_recalled_values)
macro_f1_score = np.mean(all_f1_scores)

print(f"Manually calculated Macro Precision: {macro_precision:.4f}")
print(f"Manually calculated Macro Recall: {macro_recalled_value:.4f}")
print(f"Manually calculated Macro F1-Score: {macro_f1_score:.4f}")
```

## Implement Decision Tree Classifier from Scratch

### Subtask:
Develop a `DecisionTreeClassifierScratch` class that implements a single decision tree. This class will include methods for calculating impurity (Gini or entropy), finding the best split, recursively building the tree, traversing the tree for predictions, and making predictions for an array of samples.


**Reasoning**:
To implement the `DecisionTreeClassifierScratch` class, I will define its constructor and then implement the helper methods `_gini_impurity`, `_best_split`, and the recursive tree building method `_build_tree` as per the instructions. This will lay the foundation for training the decision tree.



```python
class DecisionTreeClassifierScratch:
    def __init__(self, max_depth=10, min_samples_split=2, criterion='gini', max_features=None):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.criterion = criterion # 'gini' or 'entropy'
        self.max_features = max_features # Added max_features
        self.tree = None

    def _gini_impurity(self, y):
        if len(y) == 0: # Handle empty split
            return 0
        probabilities = np.bincount(y) / len(y)
        gini = 1 - np.sum(probabilities**2)
        return gini

    def _entropy(self, y):
        if len(y) == 0:
            return 0
        probabilities = np.bincount(y) / len(y)
        entropy = -np.sum([p * np.log2(p) for p in probabilities if p > 0])
        return entropy

    def _information_gain(self, y, y_left, y_right):
        if self.criterion == 'gini':
            parent_impurity = self._gini_impurity(y)
            left_impurity = self._gini_impurity(y_left)
            right_impurity = self._gini_impurity(y_right)
        else: # 'entropy'
            parent_impurity = self._entropy(y)
            left_impurity = self._entropy(y_left)
            right_impurity = self._entropy(y_right)

        num_samples = len(y)
        weight_left = len(y_left) / num_samples
        weight_right = len(y_right) / num_samples

        information_gain = parent_impurity - (weight_left * left_impurity + weight_right * right_impurity)
        return information_gain


    def _best_split(self, X, y):
        best_gain = -1
        best_feature_idx = None
        best_threshold = None

        n_samples, n_features = X.shape

        # Cannot split if not enough samples
        if n_samples < self.min_samples_split:
            return None, None, None, None

        # If all samples in current node belong to the same class, no split needed
        if len(np.unique(y)) == 1:
            return None, None, None, None

        # Determine features to consider for splitting
        if self.max_features is None:
            features_to_consider = np.arange(n_features)
        else:
            if isinstance(self.max_features, str):
                if self.max_features == 'sqrt':
                    num_features_to_consider = int(np.sqrt(n_features))
                elif self.max_features == 'log2':
                    num_features_to_consider = int(np.log2(n_features))
                else:
                    num_features_to_consider = n_features # Fallback for invalid string
            elif isinstance(self.max_features, (int, float)):
                if 0 < self.max_features <= 1.0:
                    num_features_to_consider = int(self.max_features * n_features)
                elif self.max_features > 1:
                    num_features_to_consider = int(self.max_features)
                else:
                    num_features_to_consider = n_features # Fallback for non-positive or invalid float
            else:
                num_features_to_consider = n_features # Fallback for invalid type

            num_features_to_consider = max(1, min(num_features_to_consider, n_features)) # Ensure valid range
            features_to_consider = np.random.choice(n_features, num_features_to_consider, replace=False)


        for feature_idx in features_to_consider: # Iterate only over selected features
            thresholds = np.unique(X[:, feature_idx])
            for threshold in thresholds:
                # Split the data
                left_indices = X[:, feature_idx] <= threshold
                right_indices = X[:, feature_idx] > threshold

                y_left, y_right = y[left_indices], y[right_indices]

                if len(y_left) == 0 or len(y_right) == 0:
                    continue

                gain = self._information_gain(y, y_left, y_right)

                if gain > best_gain:
                    best_gain = gain
                    best_feature_idx = feature_idx
                    best_threshold = threshold

        if best_gain > 0: # Only return a split if there's actual gain
            return best_feature_idx, best_threshold, best_gain, None
        return None, None, None, None

    def _build_tree(self, X, y, depth=0):
        # Get the most common class in the current node
        most_common = np.bincount(y).argmax()

        # Stopping criteria
        if depth >= self.max_depth or len(y) < self.min_samples_split or len(np.unique(y)) == 1:
            return Node(value=most_common)

        feature_idx, threshold, _, _ = self._best_split(X, y)

        # If no good split found, make it a leaf node
        if feature_idx is None:
            return Node(value=most_common)

        left_indices = X[:, feature_idx] <= threshold
        right_indices = X[:, feature_idx] > threshold

        left_child = self._build_tree(X[left_indices], y[left_indices], depth + 1)
        right_child = self._build_tree(X[right_indices], y[right_indices], depth + 1)

        return Node(feature_idx, threshold, left_child, right_child)

    def fit(self, X, y):
        self.tree = self._build_tree(X, y)

    def _traverse_tree(self, x, node):
        if node.is_leaf_node():
            return node.value

        if x[node.feature_idx] <= node.threshold:
            return self._traverse_tree(x, node.left)
        else:
            return self._traverse_tree(x, node.right)

    def predict(self, X):
        # Ensure X is numpy array for consistent indexing and operations
        X_np = np.asarray(X)
        return np.array([self._traverse_tree(x, self.tree) for x in X_np])

print("DecisionTreeClassifierScratch class defined successfully.")
```
