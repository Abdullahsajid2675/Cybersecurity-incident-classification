After EDA I observed that there are 2 features called Location and Industry included in the dataset. Based of my domain knowledge I can predict that  these feature may not be of importance when predicting attack Type.

by using a Contingency Table (Cross-tab) I verify the relation between these two feature and label seperately.


```python


# Example: df has 'location' and 'attack_type'
pd.crosstab(df['location'], df['attack_type'])
```

```python


ct = pd.crosstab(df['location'], df['attack_type'])
# Convert counts to proportions per row (location)
ct_prop = ct.div(ct.sum(axis=1), axis=0)

# Measure how uneven the distribution is
# Option 1: max proportion per location
max_prop = ct_prop.max(axis=1)
print("Max proportion per location:\n", max_prop)

# If max proportion is close to 1 → strong signal
# If max proportion is ~1/number_of_classes → weak signal
threshold = 0.6  # Example threshold
strong_signal_locations = max_prop[max_prop > threshold].index.tolist()
weak_signal_locations = max_prop[max_prop <= threshold].index.tolist()

print("Locations with weak predictive signal:", weak_signal_locations)
print("Locations with strong predictive signal:", strong_signal_locations)
```

I conclude that there is no important pattern that the model can learn from the dataset.

## Extract and Convert Attacker IP Octets

### Subtask:
Split the 'attacker_ip' column into four new columns: 'attacker_ip_oct_1', 'attacker_ip_oct_2', 'attacker_ip_oct_3', and 'attacker_ip_oct_4', by separating the IP address into its four octets, and then convert these new columns to `int32` type.


```python
df[['attacker_ip_oct_1', 'attacker_ip_oct_2', 'attacker_ip_oct_3', 'attacker_ip_oct_4']] = df['attacker_ip'].str.split('.', expand=True).astype('int32')
```


To continue with the feature engineering for IP addresses, I will now split the 'target_ip' column into four octet columns ('target_ip_oct_1', 'target_ip_oct_2', 'target_ip_oct_3', and 'target_ip_oct_4') and convert them to int32, similar to how 'attacker_ip' was handled.



```python
df[['target_ip_oct_1', 'target_ip_oct_2', 'target_ip_oct_3', 'target_ip_oct_4']] = df['target_ip'].str.split('.', expand=True).astype('int32')
```


Now that both 'attacker_ip' and 'target_ip' have been successfully split into octet-based numerical features, the original 'attacker_ip' and 'target_ip' columns are no longer needed and should be removed to clean the DataFrame and avoid redundancy.



```python
df.drop(columns=['attacker_ip', 'target_ip'], inplace=True)
print("Original 'attacker_ip' and 'target_ip' columns removed.")
```

Feature Engineering on the Timestamp Feature

```python
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")

df = df.dropna(subset=["timestamp"])

df = df.sort_values("timestamp").reset_index(drop=True)

df["year"]        = df["timestamp"].dt.year.astype("int16")
df["month"]       = df["timestamp"].dt.month.astype("int8")
df["day"]         = df["timestamp"].dt.day.astype("int8")
df["day_of_week"] = df["timestamp"].dt.weekday.astype("int8")  # 0=Mon, 6=Sun
df["hour"]        = df["timestamp"].dt.hour.astype("int8")
df["minute"]      = df["timestamp"].dt.minute.astype("int8")

#binary behavioral flags
df["is_weekend"] = (df["day_of_week"] >= 5).astype("int8")
df["is_night"]   = ((df["hour"] >= 22) | (df["hour"] <= 5)).astype("int8")
df["is_business_hours"] = ((df["hour"] >= 9) & (df["hour"] <= 17)).astype("int8")


#Cyclical Encoding (Neural Net Critical)
import numpy as np
df["hour_sin"]  = np.sin(2 * np.pi * df["hour"] / 24)
df["hour_cos"]  = np.cos(2 * np.pi * df["hour"] / 24)

df["dow_sin"]   = np.sin(2 * np.pi * df["day_of_week"] / 7)
df["dow_cos"]   = np.cos(2 * np.pi * df["day_of_week"] / 7)

df["month_sin"] = np.sin(2 * np.pi * df["month"] / 12)
df["month_cos"] = np.cos(2 * np.pi * df["month"] / 12)
```

```python
# ============================================================================
# FINAL PREPROCESSING FOR MODEL
# ============================================================================

print("="*80)
print("FINAL PREPROCESSING FOR MACHINE LEARNING")
print("="*80)

# 1. Separate target from features
print("\n1. Separating target variable...")
target = df['outcome']  # This is what we want to predict
df_features = df.drop(columns=['outcome'])

# 2. Encode target variable
from sklearn.preprocessing import LabelEncoder
le_target = LabelEncoder()
target_encoded = le_target.fit_transform(target)
print(f"   Target classes: {le_target.classes_}")
print(f"   Encoded as: {dict(zip(le_target.classes_, le_target.transform(le_target.classes_)))}")

# 3. Identify and encode categorical columns
categorical_cols = df_features.select_dtypes(include=['object']).columns.tolist()
print(f"\n2. Encoding {len(categorical_cols)} categorical columns...")

for col in categorical_cols:
    le = LabelEncoder()
    df_features[col] = le.fit_transform(df_features[col])
    print(f"   ✓ Encoded: {col}")

# 4. Remove timestamp if still present
if 'timestamp' in df_features.columns:
    df_features = df_features.drop(columns=['timestamp'])
    print("\n3. Removed timestamp column")

# 5. Combine features and target
df_final = df_features.copy()
df_final['target'] = target_encoded

# 6. Verify all columns are numeric
print("\n4. Verification:")
print(f"   Total features: {len(df_features.columns)}")
print(f"   All numeric: {df_final.select_dtypes(include=['object']).shape[1] == 0}")
print(f"   Shape: {df_final.shape}")

# 7. Save the properly preprocessed data
df_final.to_csv('preprocessed_data_final.csv', index=False)
print("\n✓ Saved: preprocessed_data_final.csv")

print("\n" + "="*80)
print("PREPROCESSING COMPLETED!")
print("="*80)
print("\nUse 'preprocessed_data_final.csv' for model training")
print("Target column: 'target' (0=Failure, 1=Success)")
print("="*80)
```
